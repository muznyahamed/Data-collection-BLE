```python
import os
import pandas as pd

def merge_csv_files(folders):
    # Create an empty list to store the dataframes
    dataframes = []
    dataframess = []


    # Iterate over the folders
    for folder in folders:
        folder_path = os.path.join('.', folder)  # Assuming the folders are in the same directory as the script

        # Find the CSV file inside the folder
        csv_files = [file for file in os.listdir(folder_path) if file.endswith('Accelerometer.csv')]
        csv_filess = [file for file in os.listdir(folder_path) if file.endswith('Gyroscope.csv')]

        if len(csv_files) == 1:
            csv_file_path = os.path.join(folder_path, csv_files[0])

            # Read the CSV file into a dataframe
            df = pd.read_csv(csv_file_path, sep='\t',usecols=lambda column: column != 'Time (s)')

            # Remove first and last 100 rows
            df = df.iloc[500:-500]

            # Add the dataframe to the list
            dataframes.append(df)
        if len(csv_filess) == 1:
            csv_file_path = os.path.join(folder_path, csv_filess[0])

            # Read the CSV file into a dataframe
            dff = pd.read_csv(csv_file_path, sep='\t',usecols=lambda column: column != 'Time (s)')

            # Remove first and last 100 rows
            dff = dff.iloc[500:-500]

            # Add the dataframe to the list
            dataframess.append(dff)

    # Merge the dataframes
    merged_df = pd.concat(dataframes, axis=0)  # Select all columns using axis=0
    merged_dff = pd.concat(dataframess, axis=0)
    # Reset the index
    merged_df.reset_index(drop=True, inplace=True)
    merged_dff.reset_index(drop=True, inplace=True)

    return merged_df,merged_dff

# Example usage
folders = ['S1_MA', 'S1_MU', 'S1_AH']
merged_df_a,merged_df_g = merge_csv_files(folders)


# Print the merged dataframe
print(merged_df_a,merged_df_g)
```

```python
import pandas as pd
import math

def split_dataframe(df, split_size, label):
    # Create an empty list to store the chunks
    chunks = []
    rounded_value = math.floor(len(df)/split_size)
    p=0

    # Split the dataframe into chunks of the specified size
    for i in range(0,rounded_value):
        chunk = df.iloc[p:p+split_size].copy()  # Create a copy of the chunk
          # Assign the specified label to the chunk
        chunks.append(chunk)
        p=p+split_size
    s = len(chunks)    
    
    return chunks , s


```

```python
import random

def suffle_pd(a,b):
    # Define the lists to shuffle
    list1 = a
    list2 = b

    # Zip the lists together
    zipped = list(zip(list1, list2))

    # Shuffle the zipped list
    random.shuffle(zipped)

    # Unzip the shuffled list
    shuffled_list1, shuffled_list2 = zip(*zipped)
    return (shuffled_list1, shuffled_list2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix

def evaluate_model(model, X_train, y_train, X_test, y_test, epochs):
    model.fit(X_train, y_train, epochs=epochs)  # Fit the model

    # Training predictions and accuracy
    train_pred = np.argmax(model.predict(X_train), axis=1)
    train_accuracy = accuracy_score(y_train, train_pred)

    # Testing predictions and accuracy
    test_pred = np.argmax(model.predict(X_test), axis=1)
    test_accuracy = accuracy_score(y_test, test_pred)

    # Confusion matrices
    train_cm = confusion_matrix(y_train, train_pred)
    test_cm = confusion_matrix(y_test, test_pred)

    # Plot confusion matrix
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.heatmap(train_cm, annot=True, cmap="Blues", fmt="d", cbar=False)
    plt.title("Confusion Matrix (Training)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    plt.subplot(1, 2, 2)
    sns.heatmap(test_cm, annot=True, cmap="Blues", fmt="d", cbar=False)
    plt.title("Confusion Matrix (Testing)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    plt.tight_layout()
    plt.show()
    
    print("Test accuracy    ",test_accuracy)

    return 
```

## CNN model

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the model architecture
model = models.Sequential()
model.add(layers.Conv1D(32, kernel_size=3, activation='relu', input_shape=(240, 3)))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Conv1D(64, kernel_size=3, activation='relu'))
model.add(layers.MaxPooling1D(pool_size=2))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
